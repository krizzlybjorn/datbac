```python
def dist_mean_std_chau():
    mean_dict = {}
    stdev_dict = {}
    distance_dict = {}
    add_df = pd.concat(all_dfs, axis=1, keys=[a[:-4] for a in participant_files])
    for participant in participant_files:
        ID = participant[:-4]
        add_df_338 = add_df[ID]
        add_df_338 = add_df_338[~add_df_338.isnull().all(1)]
        add_df_338 = add_df_338.asfreq('s')
        
        distances = list(add_df_338['distance'][add_df_338['lap'].diff() == 1])

        if len(distances) == 4:
            for i in range(len(distances)):
                if i in distance_dict:
                    distance_dict[i].append(distances[i])
                else: # triggers on first only
                    distance_dict[i] = [distances[i]]
                
    #Chauvenets criterion.
    for key in distance_dict:
        distancelist = distance_dict[key]
        meanLap = mean(distancelist)
        stdevLap = stdev(distancelist)
        N = len(distancelist)
        criterion = 1.0/(2*N)
        for i in range(N):
            if i >= len(distance_dict[key]):
                break
            d = abs(distancelist[i]-meanLap)/stdevLap
            prob = erfc(d)
            if prob < criterion:
                del distance_dict[key][i]
        mean_dict[key] = mean(distance_dict[key])
        stdev_dict[key] = stdevLap
    
    return mean_dict, stdev_dict
```

```python
# Find the mean and std distance of each lap as a percentage of the whole lap, using chauvenets
def relative_dist_mean_std_chau():
    mean_dict = {}
    stdev_dict = {}
    distance_dict = {}
    add_df = pd.concat(all_dfs, axis=1, keys=[a[:-4] for a in participant_files])
    for participant in participant_files:
        ID = participant[:-4]
        add_df_338 = add_df[ID]
        add_df_338 = add_df_338[~add_df_338.isnull().all(1)]
        add_df_338 = add_df_338.asfreq('s')
        
        distances = list(add_df_338['distance'][add_df_338['lap'].diff() == 1])
        max_distance = add_df_338['distance'].max()

        if len(distances) == 4:
            for i in range(len(distances)):
                if i in distance_dict:
                    distance_dict[i].append(distances[i] / max_distance)
                else: # triggers on first only
                    distance_dict[i] = [distances[i] / max_distance]

    #Chauvenets criterion.
    for key in distance_dict:
        distancelist = distance_dict[key]
        meanLap = mean(distancelist)
        stdevLap = stdev(distancelist)
        N = len(distancelist)
        criterion = 1.0/(2*N)
        for i in range(N):
            if i >= len(distance_dict[key]):
                break
            d = abs(distancelist[i]-meanLap)/stdevLap
            prob = erfc(d)
            if prob < criterion:
                del distance_dict[key][i]
        mean_dict[key] = mean(distance_dict[key])
        stdev_dict[key] = stdevLap

    return mean_dict, stdev_dict
```
